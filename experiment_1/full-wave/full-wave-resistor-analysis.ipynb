```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("vpp_full_wave.xlsx")
df.index += 1
df
```

```python
data = {
    100: [3.96, 3.94, 3.92, 3.96, 3.98],
    470: [1.62, 1.64, 1.64, 1.62, 1.66],
    560: [1.40, 1.42, 1.41, 1.44, 1.42],
    1000: [1.02, 1.04, 1.04, 1.02, 1.02],
    2200: [0.74, 0.74, 0.76, 0.74, 0.76],
    5600: [0.52, 0.52, 0.54, 0.56, 0.56]
    100000: [0.40, 0.38, 0.42, 0.44, 0.46]
}

stderror = []

# calculating the standard error which is standard deviation/square root of number of readings
standard_errors = {resistor: np.std(readings) / np.sqrt(len(readings)) for resistor, readings in data.items()}

for resistor, se in standard_errors.items():
    print(f"Resistor {resistor} Ohms: Standard Error = {se:.4f}")
    stderror.append(round(se, 4))


df['Standard Error'] = stderror
stderror
df
```

```python
resistor = df['Resistors (Ohms)'].values
readings = df[['Reading 1', 'Reading 2', 'Reading 3', 'Reading 4', 'Reading 5']].values
averages = df['Average'].values

plt.errorbar(resistor, averages, yerr=df['Standard Error'], fmt='o', ecolor='black', capsize=5, label='Average Readings with Standard Error', linestyle='-')

plt.xlabel('Resistor (Ohms)')
plt.ylabel('Average Reading')
plt.title('Scatter Plot of Resistor Values vs. Average Readings with Standard Error')

plt.legend()
plt.show()
```

```python
# Create the time array
t = np.linspace(0, 4 * np.pi, 1000)
plt.figure(figsize=(8, 6))

# Define the AC signal
V_ac = np.sin(t)

# Define the full-wave rectified output
V_rectified = np.abs(V_ac)

# Simulate the capacitor filter effect (exponential decay)
tau = 7  # Decay constant for the capacitor filter
V_filtered = np.zeros_like(V_rectified)

# Filtered output: charging to peaks and decaying
V_filtered[0] = V_rectified[0]
for i in range(1, len(t)):
    if V_rectified[i] > V_filtered[i-1]:
        V_filtered[i] = V_rectified[i]  # Charge up to the peak
    else:
        # Exponential decay for the capacitor discharge
        dt = t[i] - t[i-1]
        V_filtered[i] = V_filtered[i-1] * np.exp(-dt / tau)

# Plot the full-wave rectified and filtered output
plt.plot(t, V_filtered, color='black', linewidth=3, label='Rectified & Filtered')

# Plot the original AC waveform as a dashed line
plt.plot(t, V_ac, color='green', linestyle='--', label='AC Signal')

# plt.figure(figsize=8,6))
# Add labels, ticks, and legends
plt.title('Full-Wave Rectifier with Capacitor Filter', fontsize=16)
plt.xlabel('Time (s)')
plt.ylabel('Current ($I_{0}$')
plt.xticks([np.pi, 2*np.pi, 3*np.pi, 4*np.pi], [r'$\pi$', r'$2\pi$', r'$3\pi$', r'$4\pi$'], fontsize=14)
# plt.yticks([-1, 0, 1], ['-V', '0', 'V'], fontsize=14)

# Set limits for better visualization
plt.xlim([0, 4 * np.pi])
plt.ylim([-1.5, 1.5])

# Add arrows to the axes to match the original style
# plt.annotate('', xy=(4*np.pi, 0), xytext=(0, 0),
#              arrowprops=dict(facecolor='black', shrink=0.05))
# plt.annotate('', xy=(0, 1.5), xytext=(0, -1.5),
#              arrowprops=dict(facecolor='black', shrink=0.05))

plt.axhline(0, color='black', linewidth=2)

# Add a legend
plt.legend(fontsize=12)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the time array
t = np.linspace(0, 4 * np.pi, 1000)
plt.figure(figsize=(8, 6))

# Define the AC signal
V_ac = np.sin(t)

# Define the half-wave rectified output (only positive half of the sine wave)
V_rectified = np.maximum(0, V_ac)

# Simulate the capacitor filter effect (exponential decay)
tau = 7  # Decay constant for the capacitor filter
V_filtered = np.zeros_like(V_rectified)

# Filtered output: charging to peaks and decaying
V_filtered[0] = V_rectified[0]
for i in range(1, len(t)):
    if V_rectified[i] > V_filtered[i-1]:
        V_filtered[i] = V_rectified[i]  # Charge up to the peak
    else:
        # Exponential decay for the capacitor discharge
        dt = t[i] - t[i-1]
        V_filtered[i] = V_filtered[i-1] * np.exp(-dt / tau)

# Plot the half-wave rectified and filtered output
plt.plot(t, V_filtered, color='black', linewidth=3, label='Rectified & Filtered')

# Plot the original AC waveform as a dashed line
plt.plot(t, V_ac, color='green', linestyle='--', label='AC Signal')

# Add labels, ticks, and legends
plt.title('Half-Wave Rectifier with Capacitor Filter', fontsize=16)
plt.xlabel('Time (s)')
plt.ylabel('Current ($I_{0}$)')
plt.xticks([np.pi, 2*np.pi, 3*np.pi, 4*np.pi], [r'$\pi$', r'$2\pi$', r'$3\pi$', r'$4\pi$'], fontsize=14)

# Set limits for better visualization
plt.xlim([0, 4 * np.pi])
plt.ylim([-1.5, 1.5])

# Add arrows to the axes to match the original style
plt.axhline(0, color='black', linewidth=2)

# Add a legend
plt.legend(fontsize=12)

# Display the plot
plt.show()
```
